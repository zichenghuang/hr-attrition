#### How is attrition dependent on Ages

```python
fig, ax = plt.subplots(2, 1, figsize=(14, 10), sharex=False)
sns.histplot(ax=ax[0], 
             data=data, 
             x='Age', 
             hue='Attrition', 
             bins='auto',
             multiple='fill', 
#              stat='count', 
             palette='muted', 
             legend=True)
sns.boxplot(ax=ax[1], 
            data=data, 
            x='Age', 
            y='Attrition', 
            palette='muted')
ax[0].set_ylabel('Probability')
ax[0].set_title('Stacked bar plot: Age by Attrition')
ax[1].set_ylabel(None)
ax[1].set_title('Boxplot: Age by Attrition');
plt.savefig('city_development_index.png')
```

#### Covert categorical variables to numerical variables

```python
data = data.drop(columns=['JobRole'])
data.Attrition.replace(to_replace = dict(Yes = 1, No = 0), inplace = True)
data.BusinessTravel.replace(to_replace = {'Non-Travel': 1, 'Travel_Rarely': 2, 'Travel_Frequently': 3}, inplace = True)
data.Department.replace(to_replace = {'Sales': 1, 'Research & Development': 2, 'Human Resources': 3}, inplace = True)
data.EducationField.replace(to_replace = {'Life Sciences': 1, 'Medical': 2, 'Marketing': 3, 'Technical Degree': 4, 'Human Resources': 5, 'Other':6}, inplace = True)
```

```python
data.OverTime.replace(to_replace = dict(Yes = 1, No = 0), inplace = True)
```

```python
data.info()
```

```python
correlation_age = data['Attrition'].corr(data['Age'])
correlation_age = np.round(correlation_age,3)
correlation_age
```

```python
correlation_income = data['Attrition'].corr(data['MonthlyIncome'])
correlation_income = np.round(correlation_income,3)
correlation_income
```

```python
correlation_distance = data['Attrition'].corr(data['DistanceFromHome'])
correlation_distance = np.round(correlation_distance,3)
correlation_distance
```

#### heat map

```python
# drop categorical variables
data = data.drop(columns=['BusinessTravel', 
                          'Department', 
                          'EducationField',
                          'Gender',
                          'MaritalStatus',
                          'OverTime',
                          'MonthlyRate'
                        ])
```

```python
data.info()
```

```python
corr_matrix = data.corr()
```

```python
f, ax = plt.subplots(figsize=(20, 16))
sns.heatmap(corr_matrix, cmap='viridis');
```
